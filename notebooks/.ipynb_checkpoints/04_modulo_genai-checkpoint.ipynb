```python
import pandas as pd

# Ruta al dataset de sellers clusterizados (ajusta si tu path es distinto)
ruta_sellers = "../datos/procesado/sellers_clustered.csv"

df_seller = pd.read_csv(ruta_sellers)

df_seller.head()
df_seller['cluster'].value_counts()
```

```python
def construir_perfil_texto(fila):
    return (
        f"Seller con {fila['num_items']} publicaciones y {fila['categorias_unicas']} categorías. "
        f"Reputación {fila['seller_rep_score']}. "
        f"Catálogo compuesto en {fila['pct_new']:.2f} por productos nuevos, "
        f"{fila['pct_used']:.2f} usados y {fila['pct_refurbished']:.2f} reacondicionados. "
        f"Metadata incompleta en {fila['pct_ne']:.2f}. "
        f"Aplica descuentos en {fila['pct_con_descuento']:.2f} de su catálogo. "
        f"Estrategia logística: DS {fila['DS']:.2f}, FLEX {fila['FLEX']:.2f}, "
        f"FBM {fila['FBM']:.2f}, XD {fila['XD']:.2f} y otros métodos {fila['Otro']:.2f}. "
        f"Diversidad del catálogo medida por entropía {fila['entropia_categorias']:.2f} "
        f"y concentración en categoría principal de {fila['pct_categoria_moda']:.2f}."
    )
```

```python
# Crear columna de texto
df_seller['perfil_texto'] = df_seller.apply(construir_perfil_texto, axis=1)
```

```python
# Ver un par de ejemplos
df_seller[['seller_nickname', 'cluster', 'perfil_texto']].head()
```

```python
import os
os.environ["OPENAI_API_KEY"] = #"insertar API key aqui" Dado que usé una apikey personal ejecuté el notebook con ella y luego la borré en el texto para evitar guardarla en .env etc
```

```python
from openai import OpenAI
client = OpenAI()

def embed_batch(text_list):
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text_list
    )
    return [item.embedding for item in response.data]
```

```python
# Tamaño máximo de la muestra total
tam_muestra = 3000

if len(df_seller) > tam_muestra:
    df_seller_sample = (
        df_seller
        .groupby('cluster', group_keys=False)
        .apply(lambda g: g.sample(
            min(len(g), tam_muestra // df_seller['cluster'].nunique()),
            random_state=13
        ))
        .reset_index(drop=True)
    )
else:
    df_seller_sample = df_seller.copy()

df_seller_sample['cluster'].value_counts().sort_index()
```

```python
batch_size = 256  # puedes subir/bajar según comodidad
texts = df_seller_sample['perfil_texto'].tolist()
n = len(texts)

embeddings = []

for i in range(0, n, batch_size):
    batch = texts[i:i + batch_size]
    emb_batch = embed_batch(batch)
    embeddings.extend(emb_batch)
    print(f"Procesados {min(i + batch_size, n)} / {n} sellers")

len(embeddings)
```

```python
df_seller_sample['embedding'] = embeddings
df_seller_sample[['seller_nickname', 'cluster', 'perfil_texto', 'embedding']].head()
```

- En esta estapa se trabajó con una muestra balanceada de 3.000 sellers (750 por cluster) por eficiencia de cómputo y costo de embeddings. En un entorno productivo, el mismo pipeline puede escalarse fácilmente al universo completo de sellers.

```python
import numpy as np
# X: matriz de embeddings
X_emb = np.vstack(df_seller_sample['embedding'].values)
# y: clusters como target
y = df_seller_sample['cluster'].values
X_emb.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(
    X_emb, y,
    test_size=0.2,
    random_state=13,
    stratify=y
)
```

```python
clf = LogisticRegression(
    max_iter=300,
    n_jobs=-1,
    multi_class='auto'
)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

y_pred = clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification report:\n")
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(cm, cmap="Blues")

ax.set_xticks(range(len(np.unique(y))))
ax.set_yticks(range(len(np.unique(y))))
ax.set_xlabel("Predicho")
ax.set_ylabel("Real")
ax.set_title("Matriz de confusión - Clasificador sobre embeddings")

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j], ha="center", va="center", color="black", fontsize=9)

plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import matthews_corrcoef
print("MCC:", matthews_corrcoef(y_test, y_pred))
```

### Observaciones Finales

El análisis permitió construir una segmentación robusta de sellers basada en su comportamiento operativo, características de catálogo, reputación, logística y calidad de metadata. Los clusters obtenidos muestran coherencia interna y un gradiente natural de madurez comercial, lo cual valida la utilidad práctica del modelo para apoyar decisiones de onboarding, activación, crecimiento y partnerships estratégicos dentro del marketplace.

Posteriormente, se desarrolló un clasificador basado en embeddings generados por LLM, capaz de asignar un nuevo seller a uno de los clusters definidos. El clasificador alcanzó métricas sólidas tanto en datos de entrenamiento como en datos reales nunca vistos, con accuracy ≈ 0.83 y MCC > 0.82, demostrando capacidad de generalización y estabilidad incluso frente al desbalance propio del marketplace.
